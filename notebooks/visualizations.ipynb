```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
df = pd.read_csv('/home/yusufbek-murodov/Documents/myfiles/Machine learning projects/webscrapping-amazon/data/cleaned/cleaned_data.csv')
```

```python
# --- Visualization 1: Distribution of Prices ---
plt.figure(figsize=(10,6))
sns.histplot(df['price'], bins=30, kde=True)
plt.title('Distribution of Laptop Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

plt.figure(figsize=(10,6))
sns.boxplot(x='brand_name', y='price', data=df, palette='deep', hue='brand_name')
plt.title('Price Distribution by Brand')
plt.xlabel('Brand')
plt.ylabel('Price')
plt.xticks(rotation=45)
plt.show()
```

```python
# --- Visualization 2: Price vs. Features (Scatter Plots) ---
features = ['ram', 'storage']
for feature in features:
    plt.figure(figsize=(10,6))
    sns.scatterplot(x=feature, y='price', data=df, hue='cpu')
    plt.title(f'Price vs. {feature}')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.show()
```

```python
# --- Visualization 3: Brand Market Share Analysis ---
plt.figure(figsize=(8, 5))
sns.countplot(x='brand_name', data=df, palette='bright', hue='brand_name')
plt.title('Laptop Count by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(8, 5))
df['brand_name'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Brand Share of Laptops')
plt.ylabel('')
plt.show()
```

```python
# --- Visualization 4: Average Price by Brand and RAM/Storage ---
features = ['ram', 'storage']

for feature in features:
    plt.figure(figsize=(10,6))
    sns.barplot(x='brand_name', y='price', data=df, hue=feature, errorbar=None)
    plt.title(f"Average Price by Brand and {feature}")
    plt.xlabel("Brand Name")
    plt.ylabel("Price")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# --- Visualization 5: Operating System Analysis ---
plt.figure(figsize=(8,5))
sns.countplot(x='operating_system', data=df, order=df['operating_system'].value_counts().index, palette='seismic', hue='operating_system')
plt.title('Laptop Count by Operating System')
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()


plt.figure(figsize=(8,5))
df['operating_system'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Operating System Share of Laptops')
plt.ylabel('')
plt.show()
```

```python
# --- Visualization 6: Rating Analysis ---
plt.figure(figsize=(10,6))
sns.boxplot(x='brand_name', y='rating', data=df, palette='coolwarm', hue='brand_name')
plt.title("Average Distribution by Brand")
plt.xlabel("Brand")
plt.ylabel("Rating")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='brand_name', y='rating', data=df, errorbar=None, palette='coolwarm', hue='brand_name')
plt.title('Average Rating by Brand')
plt.xlabel('Brand')
plt.ylabel('Rating')
plt.xticks(rotation=45)
plt.show()
```

```python
# --- Visualization 7: Review Analysis ---
plt.figure(figsize=(10,6))
sns.scatterplot(x='reviews', y='rating', data=df, hue='brand_name', palette='viridis')
plt.title("Number of Reviews vs Rating")
plt.xlabel("Number of Reviews")
plt.ylabel("Rating")
plt.show()

plt.figure(figsize=(10,6))
sns.barplot(x='brand_name', y='reviews', data=df, errorbar=None, palette='viridis', hue='brand_name')
plt.title('Average Number of Reviews by Brand')
plt.xlabel('Brand')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
# --- Visualization 8: CPU Analysis ---
plt.figure(figsize=(10,6))
sns.countplot(y='cpu', data=df, order=df['cpu'].value_counts().index, hue='cpu', palette='coolwarm')
plt.title('Laptop Count by CPU')
plt.xlabel('Count')
plt.ylabel('CPU')
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(
    x='cpu',
    y='price',
    data=df,
    errorbar=None,
    order=df.groupby("cpu")["price"].mean().sort_values().index,
    palette='viridis',
    hue='brand_name'
)
plt.title('Average Price by CPU', fontsize=14)
plt.xlabel('CPU', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.show()
```
